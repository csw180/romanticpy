```python
import pandas as pd
```

* pandas의 read_html 에 주소를 입력하면 해당 주소에 있는 표형태의 데이터는 모두 가져온다
* 표형태의 데이터는 리스트에 담겨져 있고 리스크의 각 요소는 DataFrame 이다

```python
# pd.read_html("https://sports.news.naver.com/kbaseball/record/index?category=kbo")[0]
df = pd.read_html("https://sports.news.naver.com/kbaseball/record/index?category=kbo&year=2021")[0]
df.info()
```

```python
df = None
for  i in [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022] :
# for  i in [2010] :
  url = f"https://sports.news.naver.com/kbaseball/record/index?category=kbo&year={i}"
  tempdf = pd.read_html(url)[0]
  tempdf['년도'] = i
  if df is None :
    df = tempdf.copy()
  else :
    df = pd.concat([df,tempdf],ignore_index=True)
```
