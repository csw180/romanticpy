````python
import pandas as pd
a1 = ['송중기', 1, 92, 82, 17]
a2 = ['권보아', 1, 89, 91, 17]
a3 = ['김나현', 1, 71, 91, 18]
a4 = ['박효신', 1, 90, 78, 18]
a5 = ['김선미', 1, 83, 78, 19]
a6 = ['강민경', 1, 81, 91, 19]
col = ['이름', '반', '영어', '국어', '나이']
df1 = pd.DataFrame([a1, a2, a3, a4, a5, a6], columns=col)
df2 = df1.copy()
df3 = df1.copy()
df4 = df1.copy()
````

````python
print(df1)
df1.rank()  # 전칼럼에 대하여 rank 가 구해진다.(비숫자항목등 포함)
df1["영어점수"] = df1["영어"].rank(ascending=False)
````

````python
df2["average"] = df2["국어"].rank(ascending=False)  # 같은 점수일때: method 의 default : "average"
df2["average"] = df2["국어"].rank(ascending=False, method="min") # method : min   같은등수일때 최소값으로 
df2["average"] = df2["국어"].rank(ascending=False, method="max") # method : max   같은등수일때 최대값
df2["average"] = df2["국어"].rank(ascending=False, method="first") # method : first 순서대로 
df2["average"] = df2["국어"].rank(ascending=False, method="dense") # method : dense 등수를 채워서 촘촘히 순서대로 
df2
````

````python
df3["순번"] = df3["반"].rank(method="first")
df3
````


````python
import pandas as pd
````
````python
a1 = ['송중기', '남', 1, 79, 93]
a2 = ['김나현', '여', 1, 85, 85]
a3 = ['권보아', '여', 1, 91, 78]
a4 = ['김범수', '남', 1, 97, 98]
a5 = ['이효리', '여', 2, 93, 79]
a6 = ['박효신', '남', 2, 78, 90]
a7 = ['한예슬', '여', 2, 87, 74]
a8 = ['민경훈', '남', 2, 99, 89]
col1 = ["", "성별", "반", "국어", "영어"]
df1 = pd.DataFrame([a1, a2, a3, a4, a5, a6, a7, a8], columns=col1).set_index("")
````
