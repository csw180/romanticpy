````python
from selenium.webdriver.common.by import By

trs = driver.find_element(by=By.ID, value="americaIndex")\
            .find_elements(by=By.TAG_NAME, value="tr")

# text_box = driver.find_element(by=By.NAME, value="my-text")
# submit_button = driver.find_element(by=By.CSS_SELECTOR, value="button")
# submit_button = driver.find_element(by=By.XPATH, value='//*[@id="relkey"]/li/a')
````

````python
# head
ths = trs[0].find_elements(by=By.TAG_NAME, value="th")
head = []
for v in ths :
    head.append(v.text)
# data
datas = []
for i in range(1,len(trs)) :
  tds = trs[i].find_elements(by=By.TAG_NAME, value="td")
  data=[]
  for v in tds :
      data.append(v.text)
  datas.append(data)
driver.quit()
````

````python
import pandas as pd

df = pd.DataFrame(datas, columns=head)
df
````
