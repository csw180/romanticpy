### **Apply 실습**

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/panda-kim/pandas/main/apply01.csv")
df
```

```python
def age_s(age) :
  # print(f'age_s: type(age_s)={type(age)}')
  return str(age//10 * 10) +"대"

print(age_s(29))
```

```python
def age_s_sex(age_sex) :
  # print(f'age_s_sex: type(age_sex)={type(age_sex)}')
  return str(age_sex[0]//10 * 10) + age_sex[1][0]

print(age_s_sex((29,"남자")))
```

##### Series에 apply 가 걸리는 경우
> Series의 데이터값 하나하나가 함수로 전달됨       
> axis=0 or 1 을 넣으면 Error 발생

```python
# df["분류1"] = df["나이"].apply(lambda x : str(x//10 * 10) +"대")
df["분류1"] = df["나이"].apply(age_s)
df
```

##### DataFrame 에 apply가 걸리는 경우     
> axis=1 을 넣지 않으면 '나이' 칼럼에 순서대로 걸리고 '성별' 칼럼에 순서대로 걸리고 이런식으로 처리됨 ( axis=0 : row방향우선처리)     
> dataFrame에 axis=1을 넣어서 apply 적용하면 대상칼럼을 Series Type으로 한개의 변수값으로 함수에 전달된다.

```python
df["분류2"] = df.apply(lambda x : str(x["나이"]//10 * 10) + x["성별"][0], axis=1) # DataFrame전체로 apply 거는 방식
# df["분류2"] = df[["나이","성별"]].apply(age_s_sex, axis=1)  # 함수에서 사용할 칼럼만 DataFrame에서 선별해서 apply 거는 방식
df
```

### **GroupBy 실습**

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/panda-kim/pandas/main/groupby01.csv")
df
```

```python
df.groupby(["매출일자","상품명"], as_index=False)[["수량","판매금액"]].agg("sum") #as_index 로 groupby 칼럼이 index가 되는것을 피한다

df.groupby(["매출일자","상품명"])[["수량","판매금액"]].agg("sum").reset_index() #reset_index로 해도 같은결과
```

### **groupby 와 transform 으로 집계함수 열만들기**

```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
df = pd.read_csv("https://raw.githubusercontent.com/panda-kim/pandas/main/pokemon01.csv")
df
```

```python
# df["체력평균"] = df.groupby("타입")["체력"].agg("mean")
df["hp_m"]=df.groupby("타입")["체력"].transform("mean")
cond1 = df["체력"] >= df["hp_m"]
df["분류"] = df["타입"].mask(cond1,"강인").mask(~cond1,"허약") #"타입" 은 뒤에서 masking이 안되었을때 가져올 칼럼으로 하면 됨
df = df.drop("hp_m",axis=1)
df.to_clipboard(index=False)
```
