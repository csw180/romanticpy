### wikipedia 대한민국 인구통계 분석

````python
import pandas as pd
````

````python
url = "https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%B8%EA%B5%AC"
dfs = pd.read_html(url)  # table 이 많으면 list 형태로 리턴된다.
# print(type(dfs))
df = dfs[4]
df.head()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font="D2Coding")
````

* 년도별 추계인구

````python
plt.figure(figsize=(15,4))
# plt.xticks(rotation=60)
# sns.pointplot(x="연도 (년)", y="추계인구(명)", data=df)
sns.lineplot(x="연도 (년)", y="추계인구(명)", data=df)
````

````python
plt.figure(figsize=(15,4))
# plt.xticks(rotation=60)
# sns.pointplot(x="연도 (년)", y="출생자수(명)", data=df)
sns.lineplot(x="연도 (년)", y="출생자수(명)", data=df)
````

* 연도별 사망자수

````python
plt.figure(figsize=(15,4))
# plt.xticks(rotation=60)
# sns.pointplot(x="연도 (년)", y="출생자수(명)", data=df)
sns.lineplot(x="연도 (년)", y="사망자수(명)", data=df)
````

* 출생자 사망자를 한 그래프에 표시하기

````python
plt.figure(figsize=(15,6))
# plt.xticks(rotation=60)
sns.lineplot(x="연도 (년)", y="출생자수(명)", data=df)
# sns.lineplot(x="연도 (년)", y="사망자수(명)", data=df)
sns.lineplot(x="연도 (년)", y="사망자수(명)", color="orange",data=df)
````
